## 2.3.2 方法：写在类里面的函数，用来描述对象能做什么

```python
class Car:
    def __init__(self,color,brand):
        self.color = color
        self.brand = brand
        self.speed = 0

    def run(self,how_fast):
        self.speed = how_fast
        print(f"{self.color}的{self.brand}车子正在以{self.speed}km/h行驶")

    def stop(self):
        self.speed = 0
        print(f"{self.color}的{self.brand}车子正在停车，速度={self.speed}")

my_car = Car("红色","byd")
my_car.run(60)
my_car.stop()
```

```python
注意：方法的第一个参宿和永远是self，但不需要自己传入，python自动传入对象
```

# 2.3.3 类中方法分类

```python
访问权限 公有方法 无下划线 外部任意调用，对外接口

访问权限 保护方法 _method 仅类/子类内部使用，软性私有

访问权限 私有方法 __method 外部无法直接访问，强隐藏

参数标识 实例方法 首参 self 操作实例独有数据（90%场景）

参数标识 类方法 @classmethod，首参 cls 操作类共享属性

参数标识 静态方法 @staticmethod，无默认参数 纯工具计算，不碰任何属性

内置魔术 魔术方法 __func__ 前后双下划线 创建/打印对象时自动触发

业务用途 抽象模板方法 父类抛异常 强制子类重写业务逻辑
```

```python
1.按权限分类
1.1 公有方法（无下划线）
定义：方法名不加_，外部代码可只有调用，是对外提供的功能接口
2.保护方法（单下划线_xxx）
约定：仅限本类、子类内部使用，外部不建议调用，语法不强制拦截
3.私有方法（双下划线__xxx）
外部、子类无法访问，完全用于内部底层逻辑
```

```python
class Test:
    def __calc(self):
        return 1000

t = Test()
t.__calc()
```

```python
2. 按照第一个参数划分
2.1 实例方法：最常见
特征：第一个固定参数self，只能用实例调用
2.2 类方法@classmethod
装饰器，第一个参数cls，代表整个类，只能操作类属性，不能访问实例独有的数据
```

```python
class Car:
    wheel_count = 4  # 类属性：所有车都默认 4 个轮子

    def __init__(self, color, brand):
        self.color = color  # 实例属性：每辆车可以不一样

    @classmethod
    def print_total(cls):
        print(f"车论数：{cls.wheel_count}")


my_car = Car("红色","byd")
print(Car.wheel_count)  # 4，类属性可以直接用类名访问
print(my_car.wheel_count)
# my_car.wheel_count = 1
my_car.print_total()
```

```python
2.3 静态方法@staticmethod
特征：无默认参数，不用self和cls去调用；和类，实例都无关，只是挂靠在类里面的一个工具函数
```

```python
@staticmethod
def calc_tool(a,b):
return a+b
```

## 3.子类继承与重写
### 3.1 什么是继承？

```python
class Car:
    def __init__(self, color, brand):
        self.color = color #把传进来的color存到对象身上
        self.brand = brand  #把传进来的brand存到对象身上
        self.speed = 0

    def run(self, how_fast):
        self.speed = how_fast
        print(f"{self.color}的{self.brand}正在以{self.speed} km/h行驶")

    def stop(self):
        self.speed = 0
        print(f"{self.color}的{self.brand}停下来了")

    def mend(self):
        self.stop()


#电动车继承Car
class ElectricCar(Car):
    def __init__(self, color, brand, battery):
        super().__init__(color, brand) #调用父类的初始化，把颜色品牌线设好
        self.battery = battery  #电动车独有:电池容量
    def charge(self):
        print(f"{self.brand}正在充电,电池容量{self.battery}kwh")

ec = ElectricCar("蓝色", "特斯拉", 75)
ec.run(100)
ec.charge()
        
```

```python
# 3.2 重写方法
父类的方法不适用于子类，就在子类中重写一个同名方法，覆盖父类的方法
```

# 5.多线程编程 threading
1.什么是“并发”（一段时间内多个线程同时运行）普通代码，串行；
多线程：程序同时干多个任务，特别适合等待型任务（ai模型推理，等网络请求）


```python
import time

def task(name,seconds):
    print(f"任务{name}开始")
    time.sleep(seconds)
    print(f"任务{name}结束")
start = time.time()
task("A",2)
task("B",2)
task("C",2)
end = time.time()
print(f"串行总耗时:{end - start:.2f}")
```

```python
# 多线程版本
import time
import threading

def task(name,seconds):
    print(f"任务{name}开始")
    time.sleep(seconds)
    print(f"任务{name}结束")

start = time.time()
# 创建线程
t1 = threading.Thread(target=task, args=("A",2))
t2 = threading.Thread(target=task, args=("B",4))
t3 = threading.Thread(target=task, args=("C",3))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

end = time.time()
print(f"串行总耗时:{end - start:.2f}")
```

```python
2.线程安全与锁
多个线程同时修改同一变量可能会出问题，比如5个线程同时对一个计数器+1，结果会不对。锁：threading.Lock()]来保证一次只有一个线程能改
```

```python
import threading

count = 0
lock = threading.Lock()
def add():
    global count
    for _ in range(10000):
        lock.acquire()
        count += 1
        lock.release()
ts = [threading.Thread(target=add) for _ in range(5)]
for t in ts:
    t.start()

for t in ts:
    t.join() 
print(count)
```
